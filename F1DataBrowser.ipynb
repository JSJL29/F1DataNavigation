```python
# Assurez-vous de charger les fichiers CSV nÃ©cessaires
drivers_standings_df = pd.read_csv('data/driver_standings.csv')
races_df = pd.read_csv('data/races.csv')
drivers_df = pd.read_csv('data/drivers.csv')

# Ajouter l'annÃ©e aux standings des pilotes
drivers_standings_df['year'] = drivers_standings_df['raceId'].map(races_df.set_index('raceId')['year'])

# Prendre uniquement le dernier classement de chaque saison pour chaque pilote
last_standing_per_season = drivers_standings_df.sort_values(by=['year', 'raceId']).drop_duplicates(subset=['year', 'driverId'], keep='last')

# Additionner les points par pilote
total_points = last_standing_per_season.groupby('driverId')['points'].sum().reset_index()

# Joindre avec la table des pilotes pour obtenir les noms
drivers_df = drivers_df[['driverId', 'forename', 'surname']]
total_points = total_points.merge(drivers_df, on='driverId')

# Trier du plus grand au plus petit
top_driver = total_points.sort_values(by='points', ascending=False).iloc[0]

# Trouver tous les pilotes avec 0 point
zero_points_drivers = total_points[total_points['points'] == 0]

# Afficher les rÃ©sultats
print(f"ðŸ Le pilote avec le plus de points est {top_driver['forename']} {top_driver['surname']} avec un total de {top_driver['points']} points. ðŸ")

if zero_points_drivers.empty:
    print("âœ… Aucun pilote n'a terminÃ© avec 0 point.")
else:
    print("âš ï¸ Liste des pilotes ayant terminÃ© avec 0 point :")
    for _, row in zero_points_drivers.iterrows():
        print(f" - {row['forename']} {row['surname']}")

```

```python
# Charger les fichiers CSV
drivers_df = pd.read_csv('data/drivers.csv')
races_df = pd.read_csv('data/races.csv')
result_df = pd.read_csv('data/results.csv')
qualifying_df = pd.read_csv('data/qualifying.csv')

prenom = input("PrÃ©nom du pilote : ").capitalize()
nom = input("Nom du pilote : ").capitalize()

# Trouver l'ID de Charles Leclerc
charles_leclerc = drivers_df[(drivers_df['forename'] == prenom) & (drivers_df['surname'] == nom)]

if charles_leclerc.empty:
    print(f"âŒ {prenom} {nom} n'a pas Ã©tÃ© trouvÃ© dans la base de donnÃ©es.")
else:
    charles_id = charles_leclerc.iloc[0]['driverId']

    # Trouver les victoires de Leclerc
    race_wins = result_df[(result_df['driverId'] == charles_id) & (result_df['position'] == '1')]

    # Trouver les pÃ´les positions de Leclerc
    pole_positions = qualifying_df[(qualifying_df['driverId'] == charles_id) & (qualifying_df['position'] == 1)]

    # Nombre total de pÃ´les et de victoires
    nb_poles = len(pole_positions)
    nb_wins = len(race_wins)

    # Trouver les courses oÃ¹ il a gagnÃ© aprÃ¨s avoir obtenu la pole
    wins_from_pole = race_wins[race_wins['raceId'].isin(pole_positions['raceId'])]
    nb_wins_from_pole = len(wins_from_pole)

    # Calcul du taux de conversion
    conversion_rate = (nb_wins_from_pole / nb_poles * 100) if nb_poles > 0 else 0

    # Affichage des rÃ©sultats
    print(f"ðŸŽï¸ Nombre de pole positions de {prenom} {nom}: {nb_poles}")
    print(f"ðŸ† Nombre de courses gagnÃ©es : {nb_wins}")
    print(f"ðŸ† Nombre de victoires en partant de la pole : {nb_wins_from_pole}")
    print(f"ðŸ“Š Taux de conversion pole -> victoire : {conversion_rate:.2f}%")
```
